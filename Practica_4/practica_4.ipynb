```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
# import pandas as pd
```

```python
# Se cargan los datos de la se√±al EKG desde el archivo
data = sio.loadmat('data/EKG_data2.mat')

print(data.keys())
```

```python
# Se√±al EKG del diccionario de datos
ekg_signal = data['EKG'][0]

# Se visualiza
plt.figure(figsize=(10, 4))
plt.plot(ekg_signal)
plt.title('Se√±al EKG')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

Se puede recortar un poco mas la senal de manera que visualice mejor la dinamica

```python
# Recortar la se√±al EKG hasta 500 muestras
ekg_signal_recortada = ekg_signal[:500]

# Visualizar la se√±al recortada
plt.figure(figsize=(10, 4))
plt.plot(ekg_signal_recortada)
plt.title('Se√±al EKG Recortada')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

Mismo proceso para la senal EMG

```python
# Se cargan los datos de la se√±al EMG desde el archivo
data_EMG = sio.loadmat('data/EMG_data2.mat')

print(data_EMG.keys())
```

```python
signal_EMG = data_EMG['EMG'][0]

# Se visualiza
plt.figure(figsize=(10, 4))
plt.plot(signal_EMG)
plt.title('Se√±al EMG')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```

```python
# Desde la muestra 200 hasta la 800
emg_signal_recortada = signal_EMG[200:800]

# Visualizar la se√±al recortada
plt.figure(figsize=(10, 4))
plt.plot(emg_signal_recortada)
plt.title('Se√±al EMG Recortada')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
