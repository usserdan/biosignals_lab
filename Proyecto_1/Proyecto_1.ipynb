**Enfermedad de Parkinson**

  Es un trastorno neurodegenerativo progresivo que afecta principalmente el movimiento. Se caracteriza por la degeneraci√≥n de las neuronas dopamin√©rgicas, lo que genera una disminuci√≥n de la dopamina, un neurotransmisor basico para el control motor. El diagn√≥stico de la enfermedad de Parkinson es cl√≠nico y se basa en la evaluaci√≥n de los s√≠ntomas motores y no motores, ya que no existe una prueba definitiva para el Parkinson, el diagn√≥stico se realiza principalmente a trav√©s de la historia cl√≠nica y un examen neurol√≥gico detallado. El principal desaf√≠o para el diagn√≥stico temprano de la enfermedad de Parkinson es la ausencia de elementos espec√≠ficos que puedan identificar la enfermedad antes de la aparici√≥n de s√≠ntomas motores, ya que en las primeras etapas, los s√≠ntomas pueden ser leves y confundirse con el envejecimiento [3]. **Rutina detrend de scipy**

La rutina detrend de SciPy se utiliza para eliminar tendencias de series temporales, y permite realizar un analisis mas adecuado y precso de las fluctuaciones locales. La utilizamos para capturar patrones como la estacionalidad o ruido. La serie temporal consta de diferentes componentes, los cuales definen el valor de la serie temporal como se ve en la siguiente ecuacion:
$$ Xt = Tt+St+Rt$$  
En donde se tiene:
* La tendencia de la serie temporal  Tt:
  La tendencia es el cambio general a lo largo del tiempo, por ejemplo, un aumento o disminuci√≥n constante de los valores [9].
* La estacionalidad de la serie temporal  St:
  La estacionalidad es el resultado de los efectos estacionales, es decir, efectos que se repiten regularmente y que provocan un aumento o una disminuci√≥n con respecto a la tendencia general [9].
* La autocorrelaci√≥n entre las observaciones  Rt:
  La autocorrelaci√≥n modela c√≥mo el valor en el momento  t
  depende de los valores anteriores, es decir, c√≥mo  xt
  se correlaciona con  x(t-1), x(t‚àí2)... Esta autocorrelaci√≥n modela los cambios en la serie temporal que no se explican por la tendencia o la estacionalidad [9].

Los dos modelos disponibles en SciPy para el detrend son el modelo constante y el lineal [12].

**Modelo constante:** Se basa en eliminar la media para obtener una serie centrada en torno a cero, lo que es √∫til si la serie tiene una tendencia horizontal. Este proceso es equivalente a un ajuste por un modelo constante, donde solo se resta el valor promedio ùúá, como se muestra en la siguiente ecuacion.
$$x(t)=x(t)‚àíŒº$$

donde:
* x(t) es el valor de la serie temporal en el tiempo ùë°

Finalmente al restar la media con cada valor, se garantiza que la nueva serie tenga una media cero, y los datos se centran al rededor del origen.

**Modelo Lineal:** Se basa en ajustar a una l√≠nea recta a los datos de la serie temporal y se resta de estos [12]. La l√≠nea ajustada sigue la siguiente ecuacion:

**Filtro de Hampel**

El filtro de Hampel es un filtro no lineal que se utiliza principalmente para detectar anomalias de datos, especialmente en identificar valores at√≠picos que se desv√≠an de un comportamiento normal de series temporales. Este filtro es muy eficaz en situaciones en las que los datos contienen valores at√≠picos que podr√≠an sesgar la media y la desviaci√≥n est√°ndar [14].

El principio matematico que utiliza el filtro para saber si una muestra es at√≠pica se basa en calcular la media y la desviaci√≥n est√°ndar de los valores vecinos y si el valor se desvia mas de un numero fijo de desviaciones estandar, entonces es at√≠pico y se reemplaza por su valor medio [15].
El funcionamiento del filtro esta enfocado en como se estiman los dos parametros estadisticos, ya que la media y el desv√≠o no son estimadores robustos en presencia de at√≠picos. Por esta razon se utiliza la mediana y la Desviaci√≥n Absoluta Mediana (DAM).

* DAM: Esta estad√≠stica mide cu√°n diferentes son los valores de un conjunto de datos. Es una alternativa a la desviaci√≥n est√°ndar, ya que es menos sensible a valores extremos, proporcionando una estimaci√≥n m√°s robusta de la variabilidad [14].

**Entropia de Permutacion**

La entropia de un sistema es la tasa de generacion de nueva informacion o, equivalentemente, destruccion de vieja informacion. La entropia es una medida de la rapidez con la que informacion relacionada con el estado actual del sistema se convierte en irrelevante [18]. Por lo cual es una herramienta robusta de series temporales que proporciona una medida de cuantificaci√≥n de la complejidad de un sistema din√°mico capturando las relaciones de orden entre los valores de una serie temporal y extrayendo una distribuci√≥n de probabilidad de los patrones ordinales [19].

La entropia de Permutacion debe su nombre a Claude Shannon, que introdujo la nocion como una medida de la informacion. Entonces, dada una serie temporal Xn, 1 =< n =< T, la entropia de permutacion de Shannon se define como:

$$Hn = EP =  -‚àë{p(œÄ)log(p(œÄ))}$$

Esta ecuacion se conoce como La Entropia de Shannon de la frecuencia relativa de los patrones de orden. Me da la informacion contenida en la comparacion de los n valores consecutivos de la serie temporal x(t) [19]. Se puede ver que 0 ‚â§ ùêªùëõ ‚â§ log ùëõ! donde el l√≠mite inferior se alcanza para una secuencia de aumento o disminuci√≥n de la secuencia de valores, y el l√≠mite superior para un sistema completamente al azar donde todos las ùëõ! posibles permutaciones aparecen con la misma probabilidad.

```python
# Funcion para hacer el filtro de Hampel - Tomada de ejemplo en clase
def hampelfilter(wk, K, FilterParms):
    """
    Procedure to implement the Hampel filter
    """
    # Umbral para el filtro, tomado del primer elemento de FilterParms
    Thresh = FilterParms[0]
    ctr = wk#[K] # Copia del vector wk
    ref = np.median(wk)

    AbsDev = abs(wk - ref) # Calcula la desviaci√≥n absoluta
    MAD = 1.4826 * np.median(AbsDev) # Calcula la mediana de la desviacion absoluta
    TestVal = abs(ctr - ref)
    #if TestVal > Thresh * MAD:

    yk = ctr # Inicializa el vector de salida yk como una copia de ctr
    numero = np.sum((TestVal > Thresh * MAD)==True) # Cuenta el n√∫mero de elementos que superan el umbral

    yk[TestVal > Thresh * MAD] = ref # Reemplaza los valores que superan el umbral por la mediana


    return yk
```

```python
# Funcion proporcionada por el profesor
def shannon_entropy(x):

  p , _ = np.histogram(x,bins='fd') #'fd' (estimador de Freedman Diaconis)
  p = p/len(p) #proportion

  h = -p*np.log2(p)

  h[np.isnan(h)] = 0.0
  H = np.sum(h)
  return H
```

```python
def p_entropy(op): # Tomada de repositorio GitHub [17]
    ordinal_pat = op
    max_entropy = np.log(len(ordinal_pat))
    p = np.divide(np.array(ordinal_pat), float(sum(ordinal_pat)))
    return(shannon_entropy(p)/max_entropy)
```

```python
# Se define de dataframe
columnas = ['Sujeto', 'Estado', 'CV Left Stride', 'SD Left Stride', 'Entropy Left Stride',
            'CV Right Stride', 'SD Right Stride', 'Entropy Right Stride',
            'CV Left Swing', 'SD Left Swing', 'Entropy Left Swing',
            'CV Right Swing', 'SD Right Swing', 'Entropy Right Swing']
resultados_df = pd.DataFrame(columns=columnas)

# Funci√≥n para procesar cada archivo y extraer los √≠ndices
def procesar_archivo(grupo, archivo):
    archivo_path = os.path.join(directorio_actual, grupo, archivo)
    se√±ales = extraer_senales(archivo_path)

    # Se aplica el detrend y el filtro de Hampel
    rutina_detrend = detrend(se√±ales)
    eliminacion_dataAtipica = hampelfilter(rutina_detrend, 3, [6])

    # Lista para almacenar los resultados del archivo
    resultados = [archivo, grupo]

    # Iterar sobre las 4 se√±ales de inter√©s
    for i in range(4):
        se√±al = eliminacion_dataAtipica[:, i]

        # Se calcula tanto el CV como la SD
        cv, sd = cv_sd(se√±al)

        # Entropia de permutacion
        entropy = p_entropy(se√±al)

        # Se agregan los resultados al archivo
        resultados += [cv, sd, entropy]

    return resultados

# Se procesan todos los archivos de cada grupo y se agregan al dataframe
for grupo, archivos in grupos.items():  # Las claves son nombres de grupos y los valores son listas de archivos
    for archivo in archivos:
        resultados = procesar_archivo(grupo, archivo) # Procesa cada archivo en cada grupo y devuelve una lista con los resultados
        resultados_df.loc[len(resultados_df)] = resultados # Se agregan los resultados al dataframe

# Mostrar el DataFrame final con los resultados
resultados_df
```

# Planteamiento de las hip√≥tesis nulas y alternativas

Para continuar con el an√°lisis estad√≠stico y garantizar que las conclusiones sobre la variabilidad en el ciclo de marcha entre los grupos de estudio (Control, Parkinson, Huntington y ELA) sean v√°lidas, es fundamental formular correctamente las hip√≥tesis nula y alternativa. Estas hip√≥tesis nos permitir√°n, posteriormente, aplicar pruebas estad√≠sticas que determinen si existen diferencias significativas entre los grupos o si los datos cumplen con ciertos supuestos para aplicar pruebas param√©tricas o no param√©tricas. A continuacion se plantean las hipotesis para la prueba de Normalidad y de Homocedasticidad

**Para Normalidad:**

H0: Los datos provienen de una distribucion normal.

H1: Los datos no provienen de una distribucion normal.

**Para Homocedasticidad:**

H0: La varianza entre los grupos es constante (hay homocedasticidad).

H1: La varianza entre los grupos no es constante (hay heterocedasticidad).

**Supuesto de independencia:** Se da por aceptado en este contexto, ya que los sujetos de estudio en cada grupo provienen de diferentes poblaciones, es decir, individuos con diferentes enfermedades neurodegenerativas o sujetos sanos (control). Esto implica que no hay influencia entre los grupos, lo que asegura que los datos sean independientes entre s√≠.

# Selecci√≥n del tipo de prueba

Para verificar el **supuesto de normalidad** de los datos en cada uno de los grupos poblacionales (Control, Parkinson, Huntington y ELA), se utilizar√° la prueba de **Shapiro-Wilk**. Esta prueba es una de las m√°s recomendadas para evaluar la normalidad cuando el tama√±o de la muestra es peque√±o o moderado, debido a su alta sensibilidad y robustez. En este caso, la prueba de Shapiro-Wilk es la elecci√≥n adecuada, ya que todos los grupos tienen menos de 50 observaciones, lo que hace que otras pruebas de normalidad, como la prueba de Kolmogorov-Smirnov, no sean tan apropiadas en estos contextos de tama√±o reducido [26].

**Cantidad de muestras por grupo:**

- Grupo Control: n = 16
- Grupo Parkinson: n = 15
- Grupo Huntington: n = 20
- Grupo ELA: n = 13

Para verificar el **supuesto de homocedasticidad** se va a usar la **prueba de Levene**, puesto que es muy vers√°til y se puede aplicar a diferentes tipos de datos, ya que simplemente eval√∫a la dispersi√≥n o varianza dentro de cada grupo [27]. Dado que los grupos de estudio pueden tener varianzas muy diferentes debido a las patolog√≠as subyacentes (por ejemplo, pacientes con Huntington pueden mostrar una gran variabilidad en los patrones de marcha en comparaci√≥n con los sujetos control), la prueba de Levene permite una evaluaci√≥n robusta de si esas diferencias son lo suficientemente grandes como para ser significativas.

NORMALIDAD DE CV PARA CADA ESATADO (ENFERMEDAD Y CONTROL)

```python
alpha = 0.05  # Nivel de significancia
cv_columns = ['CV Left Stride', 'CV Right Stride', 'CV Left Swing', 'CV Right Swing']  # Columnas de CV
estados = ['Control', 'Parkinson', 'ALS', 'Huntington']  # Grupos a analizar

print("Prueba de Normalidad (Shapiro-Wilk) por estado y CV:\n")

for estado in estados:
    print(f"Estado: {estado}\n")
    for cv in cv_columns:
        datos = resultados_df[resultados_df['Estado'] == estado][cv].dropna()  # Se extraen los datos del estado y la columna CV

        # Prueba de normalidad Shapiro-Wilk
        test_stat, p_value = shapiro(datos)

        print(f"Columna {cv}: Estad√≠stico={test_stat}, p-valor={p_value}")
        if p_value < alpha:
            print(f"Se rechaza la hip√≥tesis nula de normalidad para {estado} en {cv}.\n")
        else:
            print(f"No se rechaza la hip√≥tesis nula de normalidad para {estado} en {cv}.\n")
```

En general, la normalidad no se cumple consistentemente en la mayor√≠a de las se√±ales, en donde se establece para las se√±ales de control se determina normaldiad para los datos de swing y se rechaza la hipotesis nula para stride. Tanto para las se√±ales de parkinson y de ALS se rachaza la hipotesis nula tanto para Left Stride, como para Right Swing y se establece normalidad para Ringt Stride y para Left Swing. Finalmente para Huntingtos se rechaza la hipotesis nula.

HOMOCEDASTICIDAD DE CV PARA CADA ESTADO (ENFERMEDAD Y CONTROL)

```python
# Se realiza la prueba de homocedasticidad
print("Prueba de Homocedasticidad (Levene) para los estados y CV:\n")

for cv in cv_columns:
    # Extracion los datos de los diferentes estados para las columnas CV
    datos_control = resultados_df[resultados_df['Estado'] == 'Control'][cv].dropna()
    datos_parkinson = resultados_df[resultados_df['Estado'] == 'Parkinson'][cv].dropna()
    datos_als = resultados_df[resultados_df['Estado'] == 'ALS'][cv].dropna()
    datos_huntington = resultados_df[resultados_df['Estado'] == 'Huntington'][cv].dropna()

    # Prueba de Levene
    stat_levene, p_value_levene = levene(datos_control, datos_parkinson, datos_als, datos_huntington)

    print(f"Columna {cv}: Estad√≠stico={stat_levene}, p-valor={p_value_levene}")
    if p_value_levene > alpha:
        print(f"No se rechaza la hip√≥tesis nula de homocedasticidad (varianzas iguales) en {cv}.\n")
    else:
        print(f"Se rechaza la hip√≥tesis nula de homocedasticidad (varianzas diferentes) en {cv}.\n")
```

En general, la homocedasticidad se cumple para la mayor√≠a de los datos de CV, excepto en CV Left Swing, donde las varianzas son diferentes entre los grupos. Esto sugiere que, en el an√°lisis posterior, es importante tener en cuenta la heterocedasticidad en este conjunto de datos.

NORMALIDAD DE SD PARA CADA ESTADO (ENFERMEDAD Y CONTROL)

```python
sd_columns = ['SD Left Stride', 'SD Right Stride', 'SD Left Swing', 'SD Right Swing']

print("Prueba de Normalidad (Shapiro-Wilk) por estado y SD:")

for estado in estados:
    print(f"Estado: {estado}\n")
    for sd in sd_columns:
        datos = resultados_df[resultados_df['Estado'] == estado][sd].dropna()  # Extraer los datos estado y columnas SD

        # Prueba de normalidad Shapiro-Wilk
        test_stat, p_value = shapiro(datos)

        print(f"Columna {sd}: Estad√≠stico={test_stat}, p-valor={p_value}")
        if p_value < alpha:
            print(f"Se rechaza la hip√≥tesis nula de normalidad para {estado} en {sd}.\n")
        else:
            print(f"No se rechaza la hip√≥tesis nula de normalidad para {estado} en {sd}.\n")
```

La normalidad no se cumple en la mayor√≠a de las se√±ales de desviaci√≥n est√°ndar (SD), especialmente en los grupos de Huntington y ALS, donde pr√°cticamente todas las se√±ales presentan una fuerte desviaci√≥n de la normalidad. En el grupo Control, la normalidad se cumple en las se√±ales de Swing, pero no en las de Stride, mientras que en Parkinson, solo algunas se√±ales cumplen con el supuesto de normalidad. Estos resultados sugieren que se deber√° considerar el uso de pruebas no param√©tricas para algunas de estas variables en los an√°lisis posteriores.

HOMOCEDASTICIDAD DE SD PARA CADA ESTADO (ENFERMEDAD Y CONTROL)

```python
print("Prueba de Homocedasticidad Levene para los estados y SD:")

for sd in sd_columns:
    # Exraccion de estados para las columnas SD
    datos_control = resultados_df[resultados_df['Estado'] == 'Control'][sd].dropna()
    datos_parkinson = resultados_df[resultados_df['Estado'] == 'Parkinson'][sd].dropna()
    datos_als = resultados_df[resultados_df['Estado'] == 'ALS'][sd].dropna()
    datos_huntington = resultados_df[resultados_df['Estado'] == 'Huntington'][sd].dropna()

    # Prueba de Levene para homocedasticidad
    stat_levene, p_value_levene = levene(datos_control, datos_parkinson, datos_als, datos_huntington)

    print(f"Columna {sd}: Estad√≠stico={stat_levene}, p-valor={p_value_levene}")
    if p_value_levene > alpha:
        print(f"No se rechaza la hip√≥tesis nula de homocedasticidad (varianzas iguales) en {sd}.\n")
    else:
        print(f"Se rechaza la hip√≥tesis nula de homocedasticidad (varianzas diferentes) en {sd}.\n")
```

En todas las columnas de desviaci√≥n est√°ndar (SD) evaluadas, no se rechaza la hip√≥tesis nula de homocedasticidad, lo que indica que las varianzas entre los grupos son homog√©neas en todos los conjuntos de datos.

NORMALIDAD DE ENTROPIA DE PERMUTACION PARA TODOS LOS ESTADOS (ENFERMEDAD Y CONTROL)

```python
# Columnas de entrop√≠a
entropy_columns = ['Entropy Left Stride', 'Entropy Right Stride', 'Entropy Left Swing', 'Entropy Right Swing']

print("Prueba de Normalidad (Shapiro-Wilk) por estado y entrop√≠a")

for estado in estados:
    print(f"Estado: {estado}\n")
    for entropy in entropy_columns:
        datos = resultados_df[resultados_df['Estado'] == estado][entropy].dropna()  # Extraer los datos del estado y entropia

        # Prueba de normalidad Shapiro-Wilk
        test_stat, p_value = shapiro(datos)

        print(f"Columna {entropy}: Estad√≠stico={test_stat}, p-valor={p_value}")
        if p_value < alpha:
            print(f"Se rechaza la hip√≥tesis nula de normalidad para {estado} en {entropy}.\n")
        else:
            print(f"No se rechaza la hip√≥tesis nula de normalidad para {estado} en {entropy}.\n")

```

Los resultados muestran que la normalidad no se cumple en ninguna de las columnas de entrop√≠a para todos los grupos (Control, Parkinson, ALS y Huntington). Esto sugiere que los an√°lisis posteriores que involucren la entrop√≠a de permutaci√≥n deber√°n utilizar pruebas no param√©tricas, dado que el supuesto de normalidad no se sostiene en ninguna de los conjuntos de datos evaluados.

HOMOCEDASTICIDAD DE ENTROPIA DE PERMUTACION PARA TODOS LOS ESTADOS (ENFERMEDAD Y CONTROL)

```python
print("Prueba de Homocedasticidad con la prueba Levene para los estados y entrop√≠a:")

for entropy in entropy_columns:
    # Extraccion de datos de diferentes estados para la columna de entrop√≠a
    datos_control = resultados_df[resultados_df['Estado'] == 'Control'][entropy].dropna()
    datos_parkinson = resultados_df[resultados_df['Estado'] == 'Parkinson'][entropy].dropna()
    datos_als = resultados_df[resultados_df['Estado'] == 'ALS'][entropy].dropna()
    datos_huntington = resultados_df[resultados_df['Estado'] == 'Huntington'][entropy].dropna()

    # Prueba de Levene para homocedasticidad
    stat_levene, p_value_levene = levene(datos_control, datos_parkinson, datos_als, datos_huntington)

    print(f"Columna {entropy}: Estad√≠stico={stat_levene}, p-valor={p_value_levene}")
    if p_value_levene > alpha:
        print(f"No se rechaza la hip√≥tesis nula de homocedasticidad (varianzas iguales) en {entropy}.\n")
    else:
        print(f"Se rechaza la hip√≥tesis nula de homocedasticidad (varianzas diferentes) en {entropy}.\n")
```

En todas las columnas de entrop√≠a de permutaci√≥n evaluadas, no se rechaza la hip√≥tesis de homocedasticidad, lo que sugiere que las varianzas entre los grupos son homog√©neas. Esto implica que, en t√©rminos de varianza, las comparaciones entre grupos pueden realizarse sin preocupaciones sobre diferencias significativas en la dispersi√≥n de los datos.

# **Seleccion de Prueba Parametrica o No Parametrica**

A partir de los resultados obtenidos, se observa que en la mayor√≠a de los casos los datos no siguen una distribuci√≥n normal, especialmente en las m√©tricas de entrop√≠a de permutaci√≥n y en varios casos de CV y SD, seg√∫n lo indicado por la prueba de Shapiro-Wilk. Aunque el supuesto de homocedasticidad (igualdad de varianzas) se cumpli√≥ en las m√©tricas evaluadas, la ausencia de normalidad en la mayor√≠a de las se√±ales sugiere que una prueba param√©trica como el ANOVA no ser√≠a adecuada en este contexto.

El principal motivo para optar por una prueba no param√©trica, como la ***prueba de Kruskal-Wallis***, es que esta no requiere que los datos sigan una distribuci√≥n normal [28]. Dado que la normalidad es un supuesto fundamental para las pruebas param√©tricas (como el ANOVA), su incumplimiento en este estudio justifica la elecci√≥n de m√©todos no param√©tricos. En este caso, la prueba de Kruskal-Wallis es particularmente √∫til porque permite evaluar si existen diferencias significativas a nivel estad√≠stico entre los grupos en funci√≥n de las m√©tricas evaluadas (CV, SD, y entrop√≠a de permutaci√≥n). Este enfoque es apropiado tanto para datos ordinales como continuos, siempre y cuando se trate de comparaciones entre dos o m√°s grupos independientes, como es el caso de los grupos de enfermedades neurodegenerativas frente al grupo control [29]. Adem√°s, al no depender de la igualdad de varianzas (aunque este supuesto se cumple), la prueba es vers√°til y puede entregar resultados m√°s confiables, considerando las caracter√≠sticas observadas en los datos.

**Se plantean las hipotesis para la prueba de Kruskal-Wallis:**

H0: Las distribuciones de las variables en todos los grupos son iguales (no hay diferencia en las medianas)

H1: Al menos un grupo tiene una distribucion de las variables diferente (hay diferencia en las medianas)

```python
from scipy.stats import kruskal
print("Prueba de Kruskal-Wallis para cada columna y grupo (Enfermedad y Control):\n")

# Columnas de interes
all_columns = cv_columns + sd_columns + entropy_columns

for column in all_columns:
    # Extrae los datos de los diferentes estados para la columna actual
    datos_control = resultados_df[resultados_df['Estado'] == 'Control'][column].dropna()
    datos_parkinson = resultados_df[resultados_df['Estado'] == 'Parkinson'][column].dropna()
    datos_als = resultados_df[resultados_df['Estado'] == 'ALS'][column].dropna()
    datos_huntington = resultados_df[resultados_df['Estado'] == 'Huntington'][column].dropna()

    # Prueba Kruskal-Wallis
    stat_kruskal, p_value_kruskal = kruskal(datos_control, datos_parkinson, datos_als, datos_huntington)

    print(f"Columna {column}: Estad√≠stico={stat_kruskal}, p-valor={p_value_kruskal}")
    if p_value_kruskal < alpha:
        print(f"Se rechaza la hip√≥tesis nula de igualdad de medianas para {column}.\n")
    else:
        print(f"No se rechaza la hip√≥tesis nula de igualdad de medianas para {column}.\n")
```

Los resultados indican que para las m√©tricas de coeficiente de variaci√≥n (CV) y desviaci√≥n est√°ndar (SD), las diferencias entre los grupos son altamente significativas. En todas las se√±ales evaluadas (Left Stride, Right Stride, Left Swing, Right Swing), tanto para CV como para SD, los valores p son extremadamente bajos (por debajo de 0.0001), lo que llev√≥ al rechazo de la hip√≥tesis nula en cada caso. Esto sugiere que existe una diferencia estad√≠sticamente significativa en la variabilidad y la dispersi√≥n del ciclo de marcha entre los diferentes grupos.

En cuantro a la entropia de permutacion, no se rechaz√≥ la hip√≥tesis nula para ninguna de las se√±ales de entrop√≠a, lo que indica que no hay diferencias estad√≠sticamente significativas en la complejidad de la marcha entre los grupos. Esto puede indicar que, aunque las enfermedades neurodegenerativas afecten el ciclo de marcha, el patr√≥n subyacente de complejidad en los movimientos puede mantenerse relativamente similar entre los grupos, especialmente si la entrop√≠a est√° captando la regularidad de movimientos m√°s pasivos como el balanceo.

Una posible explicaci√≥n para esta diferencia es que la entrop√≠a de permutaci√≥n, tal como se utiliz√≥ en nuestro estudio, podr√≠a no captar las sutilezas de las din√°micas a largo plazo de la marcha que el an√°lisis de Hausdorff logr√≥ identificar mediante m√©todos no lineales como el an√°lisis de fluctuaci√≥n detrendida (DFA) [21].

# Medidas que podr√≠an servir para discriminar poblaciones cl√≠nicas de los controles

Para discriminar entre las poblaciones cl√≠nicas (Parkinson, Huntington, ELA) y los controles sanos en el contexto del an√°lisis del ciclo de marcha, las m√©tricas m√°s √∫tiles ser√≠an el **coeficiente de variaci√≥n (CV) y la desviaci√≥n est√°ndar (SD)**, dado que ambas han mostrado una capacidad clara para capturar las diferencias significativas en la variabilidad y estabilidad del ciclo de marcha entre los grupos, seg√∫n tanto nuestro estudio como el art√≠culo base de Hausdorff et al [21].

El **CV** es una m√©trica clave porque refleja la variabilidad relativa en los tiempos de paso y balanceo, lo cual es crucial en el an√°lisis de la marcha en enfermedades neurodegenerativas. Tanto nuestro estudio como el de Hausdorff encontraron que los grupos afectados, especialmente aquellos con Huntington y ELA, presentaban un CV significativamente mayor en comparaci√≥n con el grupo control. En particular, el CV es √∫til para discriminar entre los grupos porque capta el deterioro progresivo del control motor que se manifiesta en un aumento de la variabilidad del tiempo de paso y balanceo. Las diferencias observadas en el CV son consistentes con las caracter√≠sticas cl√≠nicas de estas enfermedades: el Parkinson, con su rigidez y bradicinesia, genera variabilidad moderada [23]; el Huntington, con sus movimientos coreicos, muestra fluctuaciones mucho m√°s amplias [24, 25]; y el ELA, con su progresiva debilidad muscular, tambi√©n presenta aumentos significativos en la variabilidad del ciclo de marcha [21]. El CV, por lo tanto, discrimina de manera efectiva entre sujetos sanos y aquellos con enfermedades neurodegenerativas.

La **SD**, al igual que el CV, refleja la dispersi√≥n de los tiempos de paso y balanceo en relaci√≥n con su media. En nuestro estudio, la SD mostr√≥ diferencias significativas entre los grupos, lo que la convierte en otra m√©trica valiosa para discriminar las poblaciones cl√≠nicas de los controles. La SD puede ser particularmente √∫til para capturar las diferencias en la variabilidad absoluta del ciclo de marcha. Mientras que el CV proporciona una medida relativa de la variabilidad (en funci√≥n de la media), la SD refleja la dispersi√≥n absoluta de los datos, lo cual puede ser √∫til para identificar fluctuaciones extremas en las se√±ales de marcha, como las observadas en pacientes con Huntington.
