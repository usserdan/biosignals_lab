1) Implemente una función que reciba una señal de múltiples canales y épocas y calcule la Energía de promedio de cada canal.

```python
# Funcion para calcular la energia
def energiaPromedio(signal):
  canales, puntos, epocas = signal.shape

  # Se calcula la energia de cada epoca
  energia_epocas = np.sum(signal**2, axis=1)
  # Se calcula la energia promedio de cada canal
  energia_promedio = np.mean(energia_epocas, axis=1)

  return energia_promedio
```

 Calcule la energía de cada canal promediada por épocas para cada sujeto, esto para ambos grupos poblacionales. Guarde esta información en un DataFrame de columnas ‘canal’ y filas ‘#sujeto’ con los valores de energía calculados, un DataFrame para cada grupo poblacional.

```python
def energiaPromedioSujetos(path):
    # Se crea una lista de los archivos en el directorio
    archivos = os.listdir(path)
    num_sujetos = len(archivos) # Cuenta sujetos lista
    print(archivos)
    print(num_sujetos)

    # Se carga el primer archivo para obtener el número de canales
    primer_archivo = cargar_archivo(os.path.join(path, archivos[0])) # Cargo solo el priemer sujeto del listado
    num_canales = primer_archivo.shape[0]  # Tomo el shape y tomo el canal (canal, puntos, epocas)

    # Matriz para almacenar las energías promedio por canal y sujeto
    matriz_energias = np.zeros((num_sujetos, num_canales))

    # Se itera sobre cada archivo, dando el indice al elemento
    for i, archivo in enumerate(archivos):
        # el archivo .mat
        data = cargar_archivo(os.path.join(path, archivo))

        # energía promedio por canal usando tu función energiaPromedio
        energias_promedio = energiaPromedio(data)

        # Se almacena las energías promedio en la matriz
        matriz_energias[i, :] = energias_promedio

    return matriz_energias
```

```python
# Directorio de sujetos control
path_control = os.getcwd() + "/control"

# Energías promedio por canal y sujeto para el grupo control
energia_controles = energiaPromedioSujetos(path_control)

print("Energía promedio por canal para el grupo control:")
print(energia_controles)
```

```python
# Dirctorio sujetos parkinson
path_parkinson = os.getcwd() + "/parkinson"

# promedio energias por canal y sujeto para el grupo control
energia_parkinson= energiaPromedioSujetos(path_parkinson)

print("Energía promedio por canal para el grupo parkinson:")
print(energia_parkinson)
```

```python


# crea DataFrames para los grupos 'parkinson' y 'control'
df_parkinson = pd.DataFrame(energia_parkinson, columns=['canal_{}'.format(i+1) for i in range(num_canales)]) #nombra las columnas segun el numero de canales
df_control = pd.DataFrame(energia_controles, columns=['canal_{}'.format(i+1) for i in range(num_canales)])

# agrega la columna 'Sujeto' a cada DataFrame
df_parkinson['Sujeto'] = archivos_parkinson
df_control['Sujeto'] = archivos_control

# crea una columna de Orden para ordenar según el número del archivo
df_parkinson['Orden'] = df_parkinson['Sujeto'].apply(lambda x: int(re.search(r'P(\d+)', x).group(1))) #se ordenan los números en los nombres de los archivos de sujetos
df_control['Orden'] = df_control['Sujeto'].apply(lambda x: int(re.search(r'C(\d+)', x).group(1)))

# Se ordenan los DataFrames
df_parkinson = df_parkinson.sort_values('Orden').reset_index(drop=True) #Ordenar los DataFrames de forma secuencial por el número del archivo.
df_control = df_control.sort_values('Orden').reset_index(drop=True)

# Se renombra la columnda de 'Sujeto' para que sea secuencial
df_parkinson['Sujeto'] = ['Sujeto ' + str(i + 1) for i in range(len(df_parkinson))]
df_control['Sujeto'] = ['Sujeto ' + str(i + 1) for i in range(len(df_control))]

# organizados lso datos se elimina la columna orden
df_parkinson = df_parkinson.drop(columns=['Orden'])
df_control = df_control.drop(columns=['Orden'])

# Finalmente se establece la columna 'Sujeto' como índice
df_parkinson.set_index('Sujeto', inplace=True)
df_control.set_index('Sujeto', inplace=True)


```

Determine si existe diferencia estadística entre canales de cada grupo de sujetos a través de una prueba t. Compruebe los supuestos necesarios para realizar una prueba t, esto es: Normalidad de la variable,  independencia (se asume que los grupos son independientes), y homocedasticidad (use una prueba de Levene), finalmente realice la prueba t para determinar si existen diferencias entre los canales entre  grupos de sujetos. De no cumplirse los requisitos, realice entonces un análisis no paramétrico (prueba  U de Mann-Whitney). Este numeral tiene como objetivo identificar los canales que entregan  información diferencial entre pacientes Sanos y con enfermedad de Parkinson.

Normalidad

* H0 = Los datos provienen de una distribucion normal
* H1 = Los datos no provienen de una distribucion normal

Independencia

* se asume que los grupos son independientes

Homocedasticidad

* H0 = La varianza entre los grupos es constante (homocedástico).
* H1 = La varianza entre los grupos no es constante (heterocedastico).

U de Mann-Whitney

* H0 = Las dos muestras provienen de poblaciones con la misma distribución de probabilidad

* H1 = Las dos muestras provienen de poblaciones con distribuciones de probabilidad diferentes

```python
n_canales = 8  # Número de canales
alpha = 0.05  # Nivel de significancia

print("Prueba de Normalidad (Kolmogorov-Smirnov) por canal:\n")

for i in range(1, n_canales + 1): #Se itera desde el canal 1 hasta el canal 8
    columna_control = f'canal_{i}'
    columna_parkinson = f'canal_{i}'

    canal_control = df_control[columna_control] #extrae los datos por canal
    canal_parkinson = df_parkinson[columna_parkinson]

    # Prueba de normalidad para el grupo control
    test_stat_control, p_value_control = kstest(canal_control, 'norm')
    # Prueba de normalidad para el grupo parkinson
    test_stat_parkinson, p_value_parkinson = kstest(canal_parkinson, 'norm')

    print(f"Canal {i}:")
    print(f"Control: Estadístico={test_stat_control}, p-valor={p_value_control}")
    if p_value_control < alpha:
        print("Se rechaza la hipótesis nula de normalidad para el grupo control.")
    else:
        print("No se rechaza la hipótesis nula de normalidad para el grupo control.")

    print(f"Parkinson: Estadístico={test_stat_parkinson}, p-valor={p_value_parkinson}")
    if p_value_parkinson < alpha:
        print("Se rechaza la hipótesis nula de normalidad para el grupo parkinson.")
    else:
        print("No se rechaza la hipótesis nula de normalidad para el grupo parkinson.")

    print("\n")
```

Los resultados de la prueba de Kolmogorov arrojaron valores por debajo del alpha escogido, por lo cual se rechaza la hipotesis nula. De esta manera se puede decir que la distribucion tanto de los datos de control como de parkinson no siguen una distribucion normal.

A continuacion se realiza la prueba de Levene de homocedassticidad. El valor de alpha nuevamente se asume como 0.05

```python
print("Prueba de Homocedasticidad (Levene) por canal:\n")

for i in range(1, n_canales + 1):
    columna_control = f'canal_{i}'
    columna_parkinson = f'canal_{i}'

    canal_control = df_control[columna_control]
    canal_parkinson = df_parkinson[columna_parkinson]

    # Prueba de Levene para homocedasticidad
    stat_levene, p_value_levene = levene(canal_control, canal_parkinson)

    print(f"Canal {i}: Estadístico={stat_levene}, p-valor={p_value_levene}")
    if p_value_levene > alpha:
        print("No se rechaza la hipótesis nula de homocedasticidad (varianzas iguales).")
    else:
        print("Se rechaza la hipótesis nula de homocedasticidad (varianzas diferentes).")

    print("\n")
```

Los resultados de la prueba de homocedasticidad de los sujetos de Control y Parkinson en cada canal arrojaron valores de P mayores al alpha escogido, por lo cual no se rechaza la hipotesis nula y se puede decir que los datos cumplen el supuesto de que las varianzas no estan muy alejadas entre si o no tienen diferencias significativas.

De los requisitos de normalidad en los datos, homocedasticidad e independencia. No se cumple la normalidad, por lo que se realiza un analisis no parametrico con la prueba U de Mann-Whitney.

```python
print("Prueba U de Mann-Whitney por canal:\n")

for i in range(1, n_canales + 1):
    columna_control = f'canal_{i}'
    columna_parkinson = f'canal_{i}'

    canal_control = df_control[columna_control]
    canal_parkinson = df_parkinson[columna_parkinson]

    # Prueba U de Mann-Whitney para comparar las distribuciones
    stat_mannwhitney, p_value_mannwhitney = mannwhitneyu(canal_control, canal_parkinson)

    print(f"Canal {i}: Estadístico={stat_mannwhitney}, p-valor={p_value_mannwhitney}")
    if p_value_mannwhitney > alpha:
        print("Misma distribución (no se rechaza H0)")
    else:
        print("Distribución diferente (se rechaza H0)")

    print("\n")
```

La energía de una senal EEG puede interpretarse como una medida indirecta de la actividad neuronal. En términos fisiologicos, una mayor energía podría reflejar una mayor sincronización o actividad general de las neuronas en una región particular [1]. Sin embargo, la energía en sí no captura la complejidad de la senal EEG, como la frecuencia, la amplitud o las relaciones entre diferentes regiones del cerebro, que podrían ser cruciales para diferenciar estados neurologicos como el Parkinson.

Una vez realizada la prueba U de Mann-Whitney, se observa que los valores P obtenidos son mayores al alpha escogido, por lo cual no se rechaza la hipotesis nula. De esta manera, de acuerdo a las hipotesis planteadas, se puede decir que la energia de las senales en los diferentes canales entre los grupos de Control y Parkinson no presenta diferencias significativas o bastante apreciables.  Es posible que tanto los sujetos sanos como los sujetos con Parkinson muestren niveles de energía similares ya que el Parkinson afecta funciones cerebrales específicas (como el control motor [2]) que no necesariamente se traducen en grandes diferencias en la energía promedio de la senal EEG. Por lo tanto, esto podría deberse a que la energía es una medida que refleja una propiedad general de la señal y no las características más sutiles o específicas que podrían estar presentes en los patrones de oscilación cerebral.

A partir de la información obtenida, se concluye que la energía de las señales EEG no es suficiente por si sola para diferenciar de manera significativa entre sujetos sanos y sujetos con Parkinson. Esto indica que la energía promedio puede ayudar a capturar las alteraciones neurológicas relacionadas con la enfermedad, pero se deberian obtener parametros adicionales que den más claridad de los resultados. En el caso de la enfermedad de Parkinson, la fisiopatología se asocia con cambios en las redes motoras del cerebro, que pueden no siempre afectar la cantidad de energía global, sino más bien la sincronización en bandas específicas (como la beta) [3]. Entonces, el análisis de la energía promedio en canales EEG podría estar pasando por alto estas diferencias más sutiles.
