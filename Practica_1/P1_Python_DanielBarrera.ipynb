## **Reto**

Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python

def Estadisticas_Notas(notas):
    # Se crea la serie de pandas a partir del diccionario de las notas
    notas_alumnos = pd.Series(notas)

    # Se calculan las variables estadisticas y se crea una serie con esos resultados
    estadisticas = pd.Series({
        'Nota M√≠nima': notas_alumnos.min(),
        'Nota M√°xima': notas_alumnos.max(),
        'Nota Media': notas_alumnos.mean(),
        'Desviaci√≥n T√≠pica': notas_alumnos.std()
    })

    return estadisticas
```

```python
# Ejemplo de uso
alumnos = {
    'Sergio': 8.5,
    'Antonia': 7.3,
    'Daniel': 9.1,
    'Isabella': 6.8,
    'Paulina': 7.9
}

estadisticas = Estadisticas_Notas(alumnos)
estadisticas
```
