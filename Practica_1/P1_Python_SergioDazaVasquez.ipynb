Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una
serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
def notas_calculo(nota):

    notas = pd.Series(nota) #diccionario a la serie de pandas

    min_nota = notas.min() #nota minima
    max_nota = notas.max()  #nota maxima
    media_nota = notas.mean() #media de las notas
    desviacion_nota = notas.std() #la desviacion tipica

    resultado = pd.Series({'Nota m√≠nima': min_nota,'Nota m√°xima': max_nota,'Nota media': media_nota,'Desviaci√≥n t√≠pica': desviacion_nota}) #Se anexan a la serie

    return resultado
```

Verificaci√≥n funci√≥n

```python
prueba = {'Luis': 4.5, 'Daniela': 3.8, 'Juan': 4.1, 'Maria': 4.2, 'Laura': 5.0}
resultado_funcion = notas_calculo(prueba)
print(resultado_funcion)
```
