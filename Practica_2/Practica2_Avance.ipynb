Punto 1: Implemente una función en Python que permita generar una función escalón unitario, definida en
un intervalo 𝑛1≤𝑛0≤𝑛2

El escalón unitario se define como:
image.png

```python
def EscalonUni(n0, n1, n2):
  n = np.arange(n1, n2+1)
  x = (n-n0) >= 0
  return [x,n]
```

```python
x,n = EscalonUni(0, -4, 5)
plt.stem(n, x, linefmt='b', markerfmt='m')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Escalon')
plt.show
```

Punto 2: Implemente una función en Python que permita generar una función rampa, definida en un intervalo 𝑛1≤𝑛0≤𝑛2 con una pendiente m.
La función rampa se define como:
image.png

```python
def Rampa(m, n0, n1, n2):
  n = np.arange(n1, n2+1)
  x = (m*(n - n0))*((n - n0) >= 0)
  return [x, n]
```

```python
x,n = Rampa(1, 0, -4, 7)

plt.stem(n, x, linefmt='b', markerfmt='m')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Rampa')
```

* a. 𝑥1(𝑛) = 3𝛿(𝑛 + 1) + 5𝛿(𝑛 + 3) + 3𝛿(𝑛 + 2) + 3𝛿(𝑛) + 𝛿(𝑛). ¿Cómo debe ser el vectorde muestras?

```python
def ImpulsoUni(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n-n0) == 0
    return [x, n]
```

```python
x_1, n = ImpulsoUni(-1, -3, 0)
x_2, n = ImpulsoUni(-3, -3, 0)
x_3, n = ImpulsoUni(-2, -3, 0)
x_4, n = ImpulsoUni(0, -3, 0)
x_5, n = ImpulsoUni(0, -3, 0)
x = 3*x_1 + 5*x_2 + 3*x_3 + 3*x_4 + x_5

plt.stem(n, x, linefmt='b', markerfmt='m')
plt.title('Secuencia de impulsos')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.show()
```
