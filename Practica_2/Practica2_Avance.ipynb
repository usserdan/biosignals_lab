Punto 1: Implemente una funci√≥n en Python que permita generar una funci√≥n escal√≥n unitario, definida en
un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2

El escal√≥n unitario se define como:
image.png

```python
def EscalonUni(n0, n1, n2):
  n = np.arange(n1, n2+1)
  x = (n-n0) >= 0
  return [x,n]
```

```python
x,n = EscalonUni(0, -4, 5)
plt.stem(n, x, linefmt='b', markerfmt='m')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Escalon')
plt.show
```

Punto 2: Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.
La funci√≥n rampa se define como:
image.png

```python
def Rampa(m, n0, n1, n2):
  n = np.arange(n1, n2+1)
  x = (m*(n - n0))*((n - n0) >= 0)
  return [x, n]
```

```python
x,n = Rampa(1, 0, -4, 7)

plt.stem(n, x, linefmt='b', markerfmt='m')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Rampa')
```
