Punto 1: Implemente una funci√≥n en Python que permita generar una funci√≥n escal√≥n unitario, definida en
un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2

El escal√≥n unitario se define como:
image.png

```python
# A partir del codigo ejemplo del impulso, se crea la funcion de escalon
def EscalonUni(n0, n1, n2):
  n = np.arange(n1, n2+1) # Vector de muestras
  x = (n-n0) >= 0
  return [x,n]
```

```python
# Prueba de la funcion
x,n = EscalonUni(0, -4, 5)

# Se grafica
plt.stem(n, x, linefmt='b', markerfmt='m')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Funcion Escalon')
plt.show()
```

Punto 2: Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.
La funci√≥n rampa se define como:
image.png

```python
# Se define la funcion Rampa
def Rampa(m, n0, n1, n2):
  n = np.arange(n1, n2+1) # Vector de muestras
  x = np.where(n >= n0, m * (n - n0), 0) # Si se cumple que n >= n0, el valor en el array sera m * (n - n0), que es la recta de la rampa
  return [x, n]
```

```python
# Se prueba la funcion
x,n = Rampa(1, 0, -4, 7)

# Se grafica la funcion
plt.stem(n, x, linefmt='b', markerfmt='m')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Funcion Rampa')
plt.show()
```

```python
# Se genera cada tramo de funciones rampa
IntervaloUno = Rampa(1, 0, 0, 5)
IntervaloDos = Rampa(1, 5, 6, 11)
IntervaloTres = Rampa(1, 10, 12, 17)

# Se unen los tres tramos en un array
x = np.concatenate((IntervaloUno[0], IntervaloDos[0], IntervaloTres[0])) # El indice 0 es para acceder a los valores de las funciones

plt.stem(x, linefmt='b', markerfmt='m')
plt.title('Secuencia x[n] a tramos')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.show()
```

b) Realice un subplot con las dos secuencia, la original y la derivada. Describa lo que observa y el por qu√© no es exactamente la derivada del caso continuo

```python
# Se calcula la derivada
diff_x = np.diff(x)

# Se hace el subplot con la secuencia x[n] y la derivada
fig, axs = plt.subplots(2, 1, figsize=(10, 5))

axs[0].stem(x, linefmt='b', markerfmt='m')
axs[0].set_title('Secuencia x[n]')
axs[0].set_xlabel('n')
axs[0].set_ylabel('x(n)')

axs[1].stem(diff_x, linefmt='b', markerfmt='m')
axs[1].set_title('Derivada de x[n]')
axs[1].set_xlabel('n')
axs[1].set_ylabel('dx[n]')

plt.tight_layout()
plt.show()
```

En cuanto a la grafica de la derivada, la mayor√≠a de los valores son constantes y positivos, lo que corresponde a los incrementos regulares en la secuencia original. Sin embargo, se observan dos picos negativos pronunciados que coinciden con los saltos abruptos en la secuencia x[n].

En un sistema continuo, la derivada representar√≠a la tasa de cambio instant√°nea en cada punto, resultando en una funci√≥n suave. En cambio, en este caso discreto, la derivada se calcula como la diferencia entre puntos adyacentes, lo que produce valores constantes en los tramos de crecimiento uniforme y picos en los puntos de cambio brusco. Los picos negativos en la derivada indican d√≥nde la secuencia original experimenta decrementos repentinos, algo que no ser√≠a t√≠pico en una funci√≥n continua suave. Estos picos reflejan la incapacidad de una representaci√≥n discreta para capturar cambios graduales entre puntos de muestreo.
